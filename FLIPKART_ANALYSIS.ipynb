```python
def cleandata(text):
  text = str(text).lower()
  text = text.replace('!','')
  # text = re.sub('[%s]'%re.replace(string.punctuation),' ',text)
  return text
```

```python
df['summary'] = df['summary'].apply(cleandata)
```

```python
txt  = " ".join(i for i in df['summary'])
txt
```

```python
wordcloud = WordCloud().generate(txt)
plt.imshow(wordcloud)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer 
```

```python
import nltk
```

```python
nltk.download('vader_lexicon')
```

```python
sentiment = SentimentIntensityAnalyzer()
```

```python
[sentiment.polarity_scores(i) for i in df['summary']]
```

```python
df['positive'] = [sentiment.polarity_scores(i)['pos'] for i in df['summary']]
df['negative'] = [sentiment.polarity_scores(i)['neg'] for i in df['summary']]
df['neutral'] = [sentiment.polarity_scores(i)['neu'] for i in df['summary']]
```

```python
new_df = df[['summary','positive','negative','neutral']]
new_df.head(20)
```

```python
a = sum(df['positive'])
b = sum(df['negative'])
c = sum(df['neutral'])
```

```python
print('POSITIVE: ',a,'\nNEGATIVE: ',b,'\nNEUTRAL: ',c)
```
